```python
# Listas para armazenar os restaurantes e seus pratos
restaurantes = [] # Inicializa uma lista vazia para armazenar os restaurantes
pratos = [] # (Não usada) Inicializa uma lista vazia para armazenar os pratos (não é utilizada no código)

# Função para criar e exibir o menu
def exibir_menu():
    print("\nMenu:")
    print("1 - Cadastrar restaurante")
    print("2 - Cadastrar prato")
    print("3 - Fazer pedido")
    print("4 - Sair \n")


# Função para cadastrar um restaurante
def cadastrar_restaurante():
    restaurante_num = int(input("Digite o número do restaurante: "))

    # Verifica se o restaurante já existe na lista de restaurantes
    for restaurante in restaurantes:
        if restaurante[0] == restaurante_num:
            print(f"Restaurante {restaurante_num} já existe. Escolha outro número.")
            return

    # Se não existe, cadastra o restaurante
    restaurantes.append([restaurante_num, []])
    print(f"Restaurante {restaurante_num} cadastrado.")


# Função para cadastrar um prato em um restaurante
def cadastrar_prato(restaurante_num):

    restaurante_num_menor1 = restaurante_num < 1
    restaurante_num_maior_que_range = restaurante_num > len(restaurantes)

    if restaurante_num_menor1 or restaurante_num_maior_que_range:
      print("Restaurante não encontrado")
      return

    nome_prato = input("Digite o nome do prato: ")

    continuar_rodando = True
    opcao_parar = "0"

# Iniciar um loop que enquanto o usuário não inserir o valor 0, continuará pedindo Nome do prato, valor e tempo de preparo
    while (continuar_rodando):
        valor = float(input("Valor (R$): "))
        tempo = int(input("Tempo para fazer o prato (minutos): "))
        prato = [nome_prato, valor, tempo]

        # Adiciona o prato ao restaurante especificado
        restaurantes[restaurante_num - 1][1].append(prato)

        nome_prato = input("Nome do prato (ou 0 para voltar ao menu): ")

        continuar_rodando = nome_prato != opcao_parar

# Função para fazer um pedido em um restaurante
def fazer_pedido(restaurante_num):
    restaurante_existente = None

    for restaurante in restaurantes:
        if restaurante[0] == restaurante_num:
            restaurante_existente = restaurante


    if restaurante_existente == None:
        print("Restaurante não encontrado.") # Se o restaurante não existe na lista, exibe uma mensagem de erro e retorna
        return

    print(f"Restaurante {restaurante_num}:")
    pratos = restaurante_existente[1] # Obtém a lista de pratos do restaurante existente

    for i in range(len(pratos)):
        prato = pratos[i]
        print(f"{i + 1} - {prato[0]} - R$ {prato[1]:.2f} - Tempo: {prato[2]} min") # Exibe informações de cada prato

    total = 0
    tempo_total = 0
    pedido = []

    continuar_programa = True
    parar_programa = 0

    while (continuar_programa):
        escolha = int(input("Escolha um prato pelo número (ou 0 para finalizar o pedido): "))

        finalizar_pedido = 0
        escolha_igual_zero = escolha == finalizar_pedido
        escolha_menor_1 = escolha < 1
        escolha_fora_do_range = escolha > len(pratos)

        if escolha_igual_zero:
            print("Pedido finalizado! \n") # Se a escolha for igual a zero, encerra o pedido
        elif escolha_menor_1 or escolha_fora_do_range:
            print("Escolha inválida. Tente novamente.") # Se a escolha for inválida, exibe uma mensagem de erro
        else:
            prato_escolhido = pratos[escolha - 1]
            total += prato_escolhido[1] # Soma o valor do prato escolhido ao custo total
            tempo_total += prato_escolhido[2] # Soma o tempo do prato escolhido ao tempo total
            pedido.append(prato_escolhido) # Adiciona o prato escolhido ao pedido

        continuar_programa = escolha != parar_programa

    print("Seu pedido:")
    for prato in pedido:
        print(f"{prato[0]} - R$ {prato[1]:.2f} - Tempo: {prato[2]} min") # Exibe o pedido

    print(f"Seu pedido ficou R$ {total:.2f} e será entregue de {tempo_total} a {tempo_total + 10} min") # Exibe o custo total e o tempo de preparo estimado


opcao_cadastrar_restaurante = "1"
opcao_cadastrar_prato = "2"
opcao_fazer_pedido = "3"
opcao_sair = "4"

continuar_rodar_programa = True

while (continuar_rodar_programa):
    exibir_menu()
    opcao_escolhida = input("Escolha uma opção: ")

    proxima_acao_cadastrar_restaurante = opcao_escolhida == opcao_cadastrar_restaurante
    proxima_acao_cadastrar_prato = opcao_escolhida == opcao_cadastrar_prato
    proxima_acao_fazer_pedido = opcao_escolhida == opcao_fazer_pedido
    proxima_acao_sair = opcao_escolhida == opcao_sair

    if (proxima_acao_cadastrar_restaurante):
        cadastrar_restaurante()
    elif (proxima_acao_cadastrar_prato):
        restaurante_num = int(input("Digite o número do restaurante: "))
        if restaurante_num <= len(restaurantes):
            cadastrar_prato(restaurante_num) # Chama a função para cadastrar pratos em um restaurante
        else:
            print("Restaurante não encontrado.") # Se o restaurante não existe, exibe uma mensagem de erro
    elif (proxima_acao_fazer_pedido):
        restaurante_num = int(input("Digite o número do restaurante: "))
        fazer_pedido(restaurante_num) # Chama a função para fazer um pedido em um restaurante
    elif (proxima_acao_sair):
        print("Programa encerrado!")
    else:
        print("Opção inválida. Tente novamente.")

    continuar_rodar_programa = opcao_escolhida != opcao_sair
```
